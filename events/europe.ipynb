```python
# get data from .json file to a dataframe

import pandas as pd

# read json file
df = pd.read_json('events_European_Championship.json')
df.head()

df_cleaned = (
    df[['id', 'eventId', 'subEventName', 'tags', 'playerId', 'positions', 'matchId', 'eventName', 'teamId', 'matchPeriod', 'eventSec', 'subEventId']]
    .explode('tags')
    .explode('positions')
    # then we can use the .apply() method to extract the x and y coordinates
    .assign(x = lambda df: df['positions'].apply(lambda x: x['x']))
    .assign(y = lambda df: df['positions'].apply(lambda x: x['y']))
    .drop(columns=['positions'])
)


df_cleaned.head()
```
