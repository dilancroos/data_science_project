```python
# import matches_england.json file and get all the data in the nested dictionaries to a dataframe

import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read the json file
with open('matches_england.json') as json_file:
    data = json.load(json_file)

print(len(data)) # 380 matches
```

```python
# pd.json_normalize
df_normalized = pd.json_normalize(data)
print(df_normalized.columns)
print(df_normalized.shape)
df_normalized.head()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
thisFormationPlayers = []
subPlayers = []
allTeams = []
allReferees = []
for match in data:
    matchTeamsData = match['teamsData']

    for team in matchTeamsData:
        # print(team)
        teamGameStats = matchTeamsData[team]
        teamList = list(teamGameStats.values())
        teamList.insert(0, f"{match['wyId']}")
        allTeams.append(teamList)

        if teamGameStats['hasFormation'] == True:
            for formationBreak in teamGameStats['formation']:
                if formationBreak == 'substitutions': 
                    if teamGameStats['formation'][formationBreak] == 'null':
                        continue

                    else:
                        for eachSub in teamGameStats['formation'][formationBreak]:
                            vals = eachSub.values()
                            valList = list(vals)
                            valList.insert(0, f"{formationBreak}")
                            valList.insert(0, f"{match['wyId']}")
                            subPlayers.append(valList)
                        continue

                eachFormBreak = teamGameStats['formation'][formationBreak]
                if eachFormBreak == 'null':
                    continue

                else:
                    for eachPlayer in eachFormBreak:
                        vals = eachPlayer.values()
                        valList = list(vals)
                        valList.insert(0, f"{formationBreak}")
                        valList.insert(0, f"{match['wyId']}")
                        thisFormationPlayers.append(valList)

    referees = match['referees']
    refereeList = []
    refereeList.append(int(match['wyId']))
    for referee in referees:       
        refereeList.append(int(referee['refereeId']))

    allReferees.append(refereeList)
```

```python
dfPlayer = pd.DataFrame(thisFormationPlayers)
dfPlayerIndexed = dfPlayer.rename(columns={0: 'MatchId', 1: 'formation', 2: 'playerId', 3: 'ownGoals', 4: 'redCards', 5: 'goals', 6: 'yellowCards'})
dfPlayerIndexed[['MatchId', 'ownGoals', 'redCards', 'yellowCards']] = (
    dfPlayerIndexed[['MatchId', 'ownGoals', 'redCards', 'yellowCards']]
    .astype(int)
)
dfPlayerIndexed['goals'] = dfPlayerIndexed['goals'].replace('null', 0)
dfPlayerIndexed
```

```python
dfSub = pd.DataFrame(subPlayers)
dfSubIndexed = dfSub.rename(columns={0: 'MatchId', 1: 'formation', 2: 'playerIn', 3: 'playerOut', 4: 'minute'})
dfSubIndexed
```

```python
dfAllTeams = pd.DataFrame(allTeams)
dfAllTeams.drop([8], axis=1, inplace=True)
dfAllTeamsIndexed = dfAllTeams.rename(columns={0: 'MatchId', 1: 'scoreET', 2: 'coachId', 3: 'side', 4: 'teamId', 5: 'score', 6: 'scoreP', 7: 'hasFormation', 9: 'scoreHT'})  
dfAllTeamsIndexed
```

```python
dfAllMatches = pd.DataFrame(data)
dfAllMatches.drop(['teamsData'], axis=1, inplace=True)
dfAllMatches.drop(['referees'], axis=1, inplace=True)
dfAllMatches
```

```python
dfAllReferees = pd.DataFrame(allReferees)
dfAllRefereesIndexed = dfAllReferees.rename(columns={0: 'MatchId', 1: 'referee', 2: 'firstAssistant', 3: 'secondAssistant', 4: 'fourthOfficial'})
dfAllRefereesIndexed
```
